```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold   
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
%matplotlib inline
```

```python
df = pd.read_csv("data.csv")
df.drop('id',axis=1,inplace=True)
df.drop('Unnamed: 32',axis=1,inplace=True)
df['diagnosis'] = df['diagnosis'].map({'M':1,'B':0})
```

```python
df.head()
```

```python
#df.describe()
plt.hist(df['diagnosis'])
plt.title('Diagnosis (M=1 , B=0)')
plt.show()
features_mean=list(df.columns[1:11])
dfM=df[df['diagnosis'] ==1]
dfB=df[df['diagnosis'] ==0]
```

```python
plt.rcParams.update({'font.size': 8})
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(8,10))
axes = axes.ravel()
for idx,ax in enumerate(axes):
    ax.figure
    binwidth= (max(df[features_mean[idx]]) - min(df[features_mean[idx]]))/50
    ax.hist([dfM[features_mean[idx]],dfB[features_mean[idx]]], bins=np.arange(min(df[features_mean[idx]]), max(df[features_mean[idx]]) + binwidth, binwidth) , alpha=0.5,stacked=True, normed = True, label=['M','B'],color=['r','g'])
    ax.legend(loc='upper right')
    ax.set_title(features_mean[idx])
plt.tight_layout()
plt.show()
```

```python
traindf, testdf = train_test_split(df, test_size = 0.25)
def classification_model(model, data, predictors, outcome):
  arr = np.zeros((6,1))
  model.fit(data[predictors],data[outcome])
  predictions = model.predict(data[predictors])
  accuracy = metrics.accuracy_score(predictions,data[outcome])
  cm = metrics.confusion_matrix(predictions,data[outcome])
  print("Accuracy : %s" % "{0:.3%}".format(accuracy))
  kf = KFold(n_splits=5)
  error = []
  arr[0][0] = accuracy
  index = 1
  for train, test in kf.split(data):
    train_predictors = (data[predictors].iloc[train,:])
    train_target = data[outcome].iloc[train]
    model.fit(train_predictors, train_target)
    error.append(model.score(data[predictors].iloc[test,:], data[outcome].iloc[test]))  
    print("Cross-Validation Score : %s" % "{0:.3%}".format(np.mean(error)))
    arr[index][0] = np.mean(error)
    index += 1
  model.fit(data[predictors],data[outcome]) 
  sn.heatmap(cm,annot = True,fmt = 'd')
  return arr
```

```python
predictor_var = ['radius_mean','perimeter_mean','area_mean','compactness_mean','concave points_mean']
outcome_var='diagnosis'
model=LogisticRegression()
#res_log = classification_model(model,traindf,predictor_var,outcome_var)
res_log = classification_model(model,df,predictor_var,outcome_var)
```

```python
#predictor_var = ['radius_mean','perimeter_mean','area_mean','compactness_mean','concave points_mean']
predictor_var = features_mean
outcome_var='diagnosis'
model=GaussianNB()
#res_NB = classification_model(model,traindf,predictor_var,outcome_var)
res_NB = classification_model(model,df,predictor_var,outcome_var)
```

```python
predictor_var = features_mean
model = RandomForestClassifier(n_estimators=100,min_samples_split=25, max_depth=7, max_features=2)
classification_model(model, df,predictor_var,outcome_var)
featimp = pd.Series(model.feature_importances_, index=predictor_var).sort_values(ascending=False)
print(featimp)
```

```python
X = np.arange(6)
plt.figure(figsize = (20,5))
plt.plot(X,res_log,'r',label = "Logistic")
plt.plot(X,res_NB,'b',label = "Naive Bayes")
plt.legend()
```
